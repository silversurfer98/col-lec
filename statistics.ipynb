```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("BFW_pump_data.csv")
df.head()
```

```python
p = np.array(df['BFW_Pump_discharge'])
# p = np.arange(-2, 2, 0.1)
mu = np.mean(p)
var = np.var(p)
sd = np.std(p)
print("Mean : %f\nVariance : %f\nSD : %f\n" %(mu,var,sd))
```

```python
x = np.linspace(mu-5*sd,mu+5*sd,p.size)
z = (x-mu)/sd
y = (1/(np.sqrt(2*np.pi)*sd))*np.exp(-0.5*z**2)
plt.plot(x,y,'o-')
plt.axvline(mu+1.96*sd)
plt.axvline(mu-1.96*sd)
```
